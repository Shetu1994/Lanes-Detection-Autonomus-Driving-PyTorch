# Using the ConfigManager

With the ConfigManager class in config_manager.py, which you can use to get the specific filter and layer from the config.py. Here is how to use it.

```python




params = sim_config()
simulator_params = params['simulator']
# List your desired filters inside a list as a string.
filters_needed = ['RandomColorMeanFilter']
# List your desired layers inside a list as a string.
layers_needed = ['BackgroundLayer']

# Initialize ConfigManager class with simulator_params
conf_manager = ConfigManager(simulator_params)
# use get_filters method with filters_needed
conf_manager.get_filters(filters_needed)
conf_manager.get_layers(layers_needed)

# get the new params
output_params = conf_manager.params
```

This allows you to pass get the layer that you want to use and the filter that you want to use from config.py without editing the config.py.
